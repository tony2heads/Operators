```python
def fitPlaneSolve(x,y,z):
    """
    trying to solve z=A.(xy)
    """
    npts = len(z)
    A = np.array([ [sum(x*x), sum(x*y), sum(x)],
                   [sum(x*y), sum(y*y), sum(y)],
                   [sum(x),   sum(y), npts] ])
    B = np.array([ [sum(x*z), sum(y*z), sum(z)] ])
    normal = np.linalg.solve(A,B.T)
    print "Interim",np.dot(B, A), z
    nn = np.linalg.norm(normal)
    normal = normal / nn
    return normal.ravel()
```

```python
fit=fitPlaneSolve(x,y,z)
print fit
```
