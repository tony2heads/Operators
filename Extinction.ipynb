```python
#!/usr/bin/env  python
"""
Extinction model from Westerbork

"""
import math


freq=1882 #MHz


def ext(freq): #NEWSTAR offline models - clearly wrong at high freq
    f=freq/1000.
    ext=0.0057 +0.00461*f -0.000544*f*f
    return ext

def ext2(freq): #WSRT online model
    if freq<500:
        k=0.005
    elif freq > 500 and freq <1000:
        k=0.008
    elif freq > 1000 and freq <3000:
        k=0.011
    elif freq > 3000:
        k=0.015
    ext2=k
    return ext2


def gbt(freq): #used at GBT
    f=freq/1000.
    if f >52: #high freq
        tau=0.2
    elif f > 18 and f < 52: #mid freq
        tau=0.008 + math.exp(math.sqrt(f))/8000 + \
             math.exp(-0.5*(f-22.2)**2)/40.
    elif f <18: #low freq - KAT is always here
        tau=0.008 + math.exp(math.sqrt(f))/8000
    ext3=tau
    return ext3


def interpol(x,x1,x2,y1,y2):
    slope= (y2-y1)/(x2-x1)
    offset=y1
    return y1+slope*(x-x1)


def vla(freq): # table from VLAOpacityModel
    f=freq/1000.
    """
    tau=1e-3*(a+b*pwv)
    but for frq below 2GHz pwv is of no importance
    """
    if f <1.0 : x =5.27e-3
       
    if f >1.00 and f <1.25 :
        x= interpol(f, 1.0 , 1.25, 5.272,5.464)*1e-3
    if f >1.25 and f <1.5  :
        x= interpol(f, 1.25, 1.5 , 5.464, 5.579)*1e-3
    if f >1.5  and f <1.75 :
        x= interpol(f, 1.5 , 1.75, 5.579, 5.654)*1e-3
    if f >1.75 and f <2.0  :
        x= interpol(f, 1.75, 2.0 , 5.654, 5.707)*1e-3
    return x

def vlafit(freq): # fitted version of the table
     f=freq/1000.0
     if f< 2.00:
         x=0.2*f+4.6
     if f > 2.0 and f < 8.0:
         x=0.1*f + 5.5
     # rising fast near water line
     if f > 8.0 and f < 22.0:
         x=math.exp(0.08*f) +4.4 
     return x*1e-3


def kat(freq):
    """
    Just a scaled version of my VLA fit for dry atmosphere
    pressure at VLA (2124m) ~ 782 millibar
    pressure at KAT (1054m) ~ 892 millibar
    ratio of pressures KAT/VLA 1.1406
    extinction prop pressure**2 =
    1.3011
    """
    scale= 1.3011
    f=freq/1000.0
    if f< 2.00:
         x=0.2*f+4.6
    if f > 2.0 and f < 8.0:
         x=0.1*f + 5.5
     # rising fast near water line
    if f > 8.0 and f < 22.0:
         x=math.exp(0.08*f) +4.4 
    return x*1e-3*scale

def kat2(freq):
    """
    Scaled vesion of GBT model
    pressure at GBT (elev 822m) ~ 920millibar
    pressure at KAT (1054m) ~ 892 millibar
    ratio of pressures GBT/KAT 1.031390
    extinction prop pressure**2 =1.06376
    """
    scale=1.06376
    f=freq/1000.
    if f >52: #high freq
        tau=0.2
    elif f > 18 and f < 52: #mid freq
        tau=0.008 + math.exp(math.sqrt(f))/8000 + \
             math.exp(-0.5*(f-22.2)**2)/40.
    elif f <18: #low freq
        tau=0.008 + math.exp(math.sqrt(f))/8000
    ext3=tau/scale
    return ext3


"""
Pardo gives
k_dry = 2.612e-6 *(P_dry/1013)**2 * (300/T)**3.5 *(f/225)**2

P in millibar
T in K
f in GHz

k_wet  = 0.0315*(f/225)**2 *(P_wet/ 1013)*(P_dry/1013) *(300/T)**3



=========

I have also seen

gamma_dry  = 8.7e-14*p_dry**2*f**2*(0.5 +0.5/(1+(f/450)**2))*(300/T)**3.6


and also for low frequency (<<45GHz) and dry atmosphere

tau  = 1.15e-2 * (P/1013.25)**2 * (293/T)**2.85 * f**2 *K
f in GHz
P in millibar
T in Kelvin
K is an integration constant

tau in nepers == log(x1/x2)

so zenith exinction is exp(-tau)

"""
    
        


"""
The airmass function F(z) is calculated with the following model:

           1                      H   1                   H    R
   F(z) = ---- - Fz{(Rcosz)^2 + 2.- - -SQRT[(Rcosz)^2 + 2.-] - -.cosz}
          cosz                    R   H                   R    H

with :
   cosz - cos(zenith angle)
   R    - Radius of the Earth. (R = 6378160 meter)
   H    _ Scaleheight of the atmosphere, the atmosphere ia assumed to
          be homogene up to H, and empty above H. (H = 8000 meter) 
   Fz   - Model correction term, depending on the zenith angle.
            1 < z <= 83 degr : Fz = 0
           83 < z <= 86 degr : Fz = 2.454 - 20.794*cosz
           86 < z <= 88 degr : Fz = 0.0175 + 23.5*cosz

It is clear that in the zenith F(z) = 1.0, and that F(z) increases
with increasing zenith angle up to reliable values well above 20.
The model deteriorates heavily at angles between 88 and 90 degree.
"""
def airmass(z):
    # zenith angle in degreees
    cosz=math.cos(math.radians(z))
    secM1=1.0/cosz -1.0  # secant -1 
    return 1.0 + secM1 * (0.9981833 - secM1 * (0.002875 + (0.0008083 * secM1)))

def airmassW(z):
    R=6378160.0 # radius of earth /m
    h=8000.0    # scale height of atmosphere
    #WSRT formula
    #Fz is a correction factor
    cosz=math.cos(math.radians(z))
    if z >86.0 :
        Fz=0.0175 + 23.5*cosz
    if z < 86.0 and z > 83.0 :
        Fz = 2.454 - 20.794*cosz
    if z < 83.0:
        Fz=0.0
    rc= R*cosz
    return 1./cosz - Fz*((rc**2  + 2*h/R - math.sqrt((rc**2 +2*h/R))/h  -rc/h ))

def airmassR(z):
    """
    Rohlfs formula Tools Of Radio Astronomy (1986)
    only valid at elevation > 11 deg 
    """
    cosz=math.cos(math.radians(z))
    secz=1/cosz
    x=  -0.0045 + 1.00672*secz - 0.002234* secz**2 - 0.0006247*secz**3
    return x
                  

    
"""
GBT quotes attenuation of about 0.008 neper at low frequency
that is about 0.8% attenuation

VLA quotes Ulvestad as saying Tatm  = 257.9 + 0.445*(Tsurface - 273.15) K

so if Tsurface = 30C, = 303.15K  Tatm = 271K


"""
if __name__ == '__main__':

    print "WSRT models"
    print "at freq %.1fMHz zenith extinction expected is %.5f" %(freq,ext(freq))
    print "at freq %.1fMHz zenith extinction expected is %.5f" %(freq,ext2(freq))
    print "GBT model"
    print "at freq %.1fMHz zenith extinction expected is %.5f" %(freq,gbt(freq))
    print "KAT model"
    print "at freq %.1fMHz zenith extinction expected is %.5f" %(freq,kat(freq))
    print "at freq %.1fMHz zenith extinction expected is %.5f" %(freq,kat2(freq))
                                                                 
    print "VLA model"
    print "at freq %.1fMHz zenith extinction expected is %.5f" %(freq,vla(freq))
    print "at freq %.1fMHz zenith extinction expected is %.5f" %(freq,vlafit(freq))
```
