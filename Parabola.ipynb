To get plane parallel signal to arrive at a point (the focus) where all waves add up in phase we reflect from a parabola. This shape is then defined by being equal distance from a point and a straight line. Radio waves from space are effectively plane parallel waves.

parabola has the form in polar coordinates

$r = \frac{2a}{1+cos(\theta)}$

Note that this does not depend on anything else (wavelength etc.).


#Plot a parabola

```python
import numpy as np
import matplotlib.pyplot as plt

%pylab inline
pylab.figure(figsize=(12,12)) #make it a sensible size

ax = plt.subplot(111, polar=True) # one polar plot 
ax.set_theta_offset(-1*np.pi/2) #set it the right way up for a telescope

a=3  #focallength
anglemax=1.8
th =arange(-anglemax,anglemax,0.04) #  radians
r=2*a/(1+np.cos(th))  # THE PARABOLA
l=2*a/np.cos(th)      # THE LINE


ax.plot(th,r,label='parabola')
ax.plot(th,l,label='line')
plt.annotate("focus",(0,0),arrowprops=dict(arrowstyle='->'), xytext=(-7,-2*a))
ax.set_rmax(3*a) # stop the line and parabola, as they go to infinity if you let then
plt.show()
    
    
```
