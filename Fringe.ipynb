#Theory

Any Azimuth and elevation can be converted to an hour angle (H) and declination (D).

Consider a source at hour angle and declination $(h,\delta)$ and a baseline fixed on the ground of length B pointing towards a fixed direction (H,D) on the sky

As time increases only $h$ changes at the rate of $2\pi$ radians per sidereal day

If the angle between the basline the baseline and the source is $\theta$ then you can show by spherical trignometry that

$$\cos(\theta) =  sin(\delta)sin(D) -cos(\delta) cos(D) cos(H-h)$$

the **delay** in seconds 

$$\tau = \frac{B cos(\theta)}{c}$$

The **fringe rate** is the rate of change of delay (no units)

$$\dot{\tau}  =  \frac{B}{c} \frac{d cos(\theta)}{dt}  = -\frac{B sin(\theta)}{c} \frac{d\theta}{dt}$$

OR equivalently

$$= \frac{B}{c} ( cos(\delta)cos(D) sin (H-h) \frac{dh}{dt}) $$



#Special cases:

at a pole $cos(\delta) = 0 $ so the fringe rate is zero

when the source is exactly along the baseline direction $\theta =0$ so delay  = $\frac{B}{c}$ and fringe rate =0

If the baseline points north-south $cos(D) =0 $ so the fringe rate is always zero  **( a very bad idea)**

The maximum possible fringe rate is when H and h are orthogonal to each other so $sin(H-h) = 1$

#Note
delay and fringe rate are defined for a point $(h,\delta)$ on the sky. Another nearby source (within the telescope beam) will have different delay and fringe rate. Also note that when delay is a maximum or mininum the fringe rate is zero and vice-versa.

# Fringe frequency

Fringe frequency is fringe rate divided by wavelength 



If there are 2 nearby sources with fringe rates $f+1$ and $f_2$ they will have a beat period $t = \frac{1} {f_1 - f_2}$

# U,V,W

If we measure everything in wavelengths for a baseline length $b = \frac{B}{\lambda}$
##towards the source
$w = b cos(\theta)$ 

the two other components of the baseline can be diveded into a north-south and east-west component. If the angle towards north is $\sigma$
## east
$u = -b sin(\theta) sin(\sigma) = b cos(D) sin (h-H) $ 
##north
$v =  b sin(\theta) cos(\sigma) = b(cos(\delta) sin(D) - sin(\delta) cos(D) sin (h-H))$


```python
"""
Example
pick a baseline direction
pick a declination and change hour angle
"""
import numpy as np
#import matplotlib as plt

B=200 # metres
l=0.18 # metres
sidereal = 2*np.pi/(23.9344696*86400) #radians per seconds
c=2.998e8 # m/s

b=B/l
# baseline
H=np.radians(17.)
D=np.radians(61.)
# source
dec=np.radians(-41.)
#hae=np.radians(-90.)
h=np.arange(-np.pi,np.pi,0.05)
#print H,D,dec,hae
print np.cos(D),np.sin(D),np.cos(dec),np.sin(dec),b
```

```python
tau=B*(np.sin(D)*np.sin(dec) - np.cos(D)*np.cos(dec)*np.cos(H-h))/c
fr=B*sidereal *(np.cos(D)*np.cos(dec)*np.sin(H-h))/c
%pylab inline
pylab.figure(figsize=(10,10)) #make it a sensible size

pylab.plot(h,tau,label='Delay')
pylab.plot(h,fr*1e6,label=r' $10^6 \times $ fringe rate')
pylab.grid()
pylab.legend()
pylab.show()
```

```python
def uvw(D,dec,h,H,b):
    u=b*np.cos(D)*np.sin(h-H)
    v=b*(np.cos(dec)*np.sin(D) + np.sin(dec)*np.cos(D)*np.cos(h-H))
    w=b*(np.sin(dec)*np.sin(D) - np.cos(dec)*np.cos(D)*np.cos(h-H))
    return (u,v,w)
        
```

```python
#b=B/l
pylab.figure(figsize=(10,10)) #make it a sensible size

print D,dec,h[0],H,b
u,v,w=uvw(D,dec,h,H,b)

pylab.plot(h,u,label='u against h')
pylab.plot(h,v,label='v against h')


pylab.plot(h,w,label='w against h')
combined=np.sqrt(u**2 + v**2)
#pylab.plot(h,combined,marker='+',label='The combination')
bl= b*np.ones(len(h))
pylab.plot(h,bl,marker='x',label='The baseline')
pylab.grid()
pylab.legend()
pylab.show()
#print np.sqrt(combined**2+w**2)/b
```

```python
pylab.figure(figsize=(12,12)) #make it a sensible size
pylab.plot(u,v,label='(u,v) track')
pylab.axis('equal')
pylab.grid()
pylab.legend()
pylab.show()
```

#Comments

* u,v tracks are alway some sort of ellipse (circle to straight line), but not necessarily centred at (0,0)

* If the baseline is exactly east-west they are centred at (0,0)

* The 'u' component does not depend on declination, but the 'v' does.

* If you track symmetrically around hour angle you have a symmetry about the v axis


```python
dec=0.0
u,v,w=uvw(D,dec,h,H,b)
pylab.figure(figsize=(12,12)) #make it a sensible size
pylab.plot(u,v,label='(u,v) track')
pylab.axis('equal')
pylab.grid()
pylab.legend()
pylab.show()
```

If declination is zero the 'v' collapses and the tracks are straight lines, but not nessarily containing 0,0
