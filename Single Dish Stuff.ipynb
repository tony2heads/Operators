#How it works


## Focus box

* Signals are reflected off the parabolic dish and come into the dipoles at the focus (inside the Horn) http://en.wikipedia.org/wiki/Feed_horn
* RF sinals in the dipole are symmetrical about zero (no DC component) but we want a ground with signal off it, so it is converted in the OrthoMode Transducer http://en.wikipedia.org/wiki/Orthomode_transducer   and Balun http://en.wikipedia.org/wiki/Balun (in one unit)
* Calibrated noise can be injected
* There is a low noise amplifier (LNA) inside the cryostat http://en.wikipedia.org/wiki/Cryostat. This is kept at 75-80K to reduce thermal noise in the system.
* Once our Radio Frequency (RF) signal level is high enough it goes on to another amplifier with a filter (to reduce unwanted signals) in stage 3

## Pedstal
* More amplification, with attenuators that we can set to get the right levels, and more filtering
* Modulate a laser with the RF signal and send it to Losberg via the ASC

# Losberg
* Convert the optical signal back to RF and change the 1.2-1.9GHz down in frequency, so that the samplers can convert it to a digital signal.
* The digital signals go into the Digital Back End (DBE)

# Tipping curves

* The atmosphere is **almost** completely transparent at L-band (1-2GHz). That is why it is used for GPS, GSM phones etc.
* BUT if it absorbs 1% of the signal it must also emit 1% of its own temperature (about 280-300K)
* The KAT-7 site is high enough that at zenith (straight up) the atmospheric absorption is less than 1% (typically about 0.7%)
* By tipping the antenna we go through more atmosphere, and if we can calculate the effect of the emission (from the change) we can calculate the absorption and the receiver noise without the atmosphere **as long as we avoid strong sources**

# Noise measurement


* One measurement of receiver +  'cold' sky (no Sun, no Moon, no Milky Way, no strong sources , etc.)
  * this should also be at fairly high elevation
* ditto with noise source on
* One measurement of receiver + 'hot box' (absorber at external temperature)
* ditto with noise source on

* When we know the equivalent temperatures of the 'cold' sky (typically about 5K, 3K from the universe and 2K from our own atmosphere) and the hot box (about 300K) we can work out the the noise source levels.

* We do the noise source on with both systems to check that the power increase is the same, so that we know the power is measured in the linear regime

# This is a typical type of output curve for a detector. 

* Red is an ideal linear response (what you would like)
* Black is what you get. Above 2.5 units input you are maxed out. Any change in input makes almost no difference to the output and the slope gets flatter as go up.
* Note that both responses have a small offset from zero. Our dector reports a non-zero value even when there is no input. Most systems have this, as they cannot report a value below zero.
* The linear  part is from an input of 0 to 0.5
* At very low input levels digitizers don't work properly, so in this case we would select a input of 0.1-0.5

##What auto attenuate does:

* Pick a level that is high enough for the digital sampler in stage 7 to work properly
* BUT pick a level that is low enough for all the other system to stay linear



The curve above is arctan(x) +0.1
so no matter how big the input is the output cannot go above $  \frac{\pi}{2} +0.1$

# You must Autoattenuate when

* You go to a very strong source (satellite, sun) on that source!
* You change correlator mode
* You change frequency by more than a few MHz. (There is a slight change of power level over the receiver's band). OH monitoring does not need it when changing from observing 1665 to 1667MHz.
* Somebody has changed a receiver or cabling
* Power levels are reporting errors

## You don't need to autoattenuate when

* The setup is the same as the last, and all power levels are good, and you are not going to a strong source
