```python
def tophat(x,w):    """    top hat function of width w    vector x    """    d=w/2. # half width    f=zeros(len(x)) # fill vector f with zeros    for i in range(len(x)):        if abs(x[i]) < d:            f[i]=.1    return f
```

```python
def triangle(x,w):    """    triangle function ramps up to peak at origin    width w    """    d=w/2.    f=zeros(len(x))    for i in range(len(x)):        if abs(x[i]) <d:            if x[i] <0:                f[i]= 1 +x[i]/d            if x[i] > 0:                f[i] = 1 -x[i]/d    return f
```

```python
def spike(x,w):    """    spike at position w in an otherwise blank input    """    f=zeros(len(x))    nn=argmin(abs(x-w)) # find index where value of x closest to w    f[nn]=0.1    return f
```

```python
x=arange(-10*pi,10*pi,0.001) # a vector of linearly increasing valuesxmin=min(x)xmax=max(x)scale=(10*pi)/len(x) # just to make plots reasonable
```

```python
f=cos(x)pylab.figure(1,figsize=(10,10))pylab.plot(x,scale*real(fftshift(fft(f))),label='real fft')pylab.plot(x,scale*imag(fftshift(fft(f))),label='imag fft')pylab.plot(x,f,label='cosine')pylab.legend()pylab.show()
```
